```python
import openai
import pandas as pd
```

```python
REFERENCE_FILE = "../bird-data-train/train.json"

def get_embeddings_ada(df, column_name):
    embeddings = []
    for text in df[column_name]:
        # Call the OpenAI API to get the embedding using the "ada-002" model
        response = openai.Embedding.create(input=text, engine="text-embedding-3-large")
        # Extract the embedding vector
        embedding = response['data'][0]['embedding']
        embeddings.append(embedding)
    return embeddings
```

```python
df_question = pd.read_json(REFERENCE_FILE)
embeddings = get_embeddings_ada(df_question, 'question')
df_question['embeddings'] = embeddings

df_question[['question', 'embeddings']].to_csv('BIRD_train_embeddings.csv', index=False)
```
